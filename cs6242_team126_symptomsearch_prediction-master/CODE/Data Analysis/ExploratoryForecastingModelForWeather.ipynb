```
test_range = int(merged_df_filt_fips.shape[0]*.2)
merged_df_filt_fips_model_data = merged_df_filt_fips[['avg_temperature_air_2m_f','avg_humidity_relative_2m_pct']]
train_data = merged_df_filt_fips_model_data[1:len(merged_df_filt_fips_model_data)-test_range]
test_data = merged_df_filt_fips_model_data[len(merged_df_filt_fips_model_data)-test_range:]
```

```
from statsmodels.tsa.ar_model import AR

from statsmodels.tsa.arima_model import ARIMA
```

```

model = AR(train_data[['avg_temperature_air_2m_f']])
model_fitted = model.fit(method='mle', maxlag=7, disp=1, maxiter=1000)
```

```
print('lag value: %s' % model_fitted.k_ar)


print('coefficients:\n %s' % model_fitted.params)
```

```
predictions = model_fitted.predict(
    start=len(train_data), 
    end=len(train_data) + len(test_data)-1, 
    dynamic=False)

# create a comparison dataframe
compare_df = pd.concat(
    [merged_df_filt_fips['avg_temperature_air_2m_f'].tail(test_range),
    predictions], axis=1).rename(
    columns={'stationary': 'actual', 0:'predicted'})

#plot the two values
compare_df.plot()
```

```
test_range = int(merged_df_filt_fips.shape[0]*.2)
merged_df_filt_fips_model_data = merged_df_filt_fips[['avg_temperature_air_2m_f','avg_humidity_relative_2m_pct']]
train_data = merged_df_filt_fips_model_data[1:len(merged_df_filt_fips_model_data)-test_range]
test_data = merged_df_filt_fips_model_data[len(merged_df_filt_fips_model_data)-test_range:]
train, test =  [merged_df_filt_fips['avg_temperature_air_2m_f'][1:len(merged_df_filt_fips_model_data)-test_range],  merged_df_filt_fips['avg_temperature_air_2m_f'][len(merged_df_filt_fips_model_data)-test_range:]]
# test
#https://machinelearningmastery.com/make-sample-forecasts-arima-python/ used logic from here
history = [x for x in train]
predictions = list()
actual_list = []
predicted_list = []
for t in range(len(test)):
	model = ARIMA(history, order=(7,1,0))
	model_fit = model.fit(disp=0)
	output = model_fit.forecast()
	yhat = output[0]
	predictions.append(yhat)
	obs = test[t]
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
	actual_list.append(obs)
	predicted_list.append(yhat)
```

```

from sklearn.metrics import r2_score 
r2_score(actual_list, predicted_list)
```

```

from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(actual_list, predicted_list)
```

```

from matplotlib import pyplot
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
